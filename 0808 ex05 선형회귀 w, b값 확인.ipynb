### 목표 설정

- 간단한 직선을 그어 W(가중치) / B(편향) 확인

```python
# 경고창 무시

import warnings
warnings.filterwarnings("ignore")
```

```python
# 필요한 라이브러리

import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt 
```

```python
# 선형회귀 모델 

from sklearn.linear_model import LinearRegression
```

```python
# 데이터 준비

X = range(1, 6)
y = [2, 4, 6, 8, 10]
```

```python
# 그래프 상으로 X데이터와 y데이터 확인

plt.plot(X, y, marker="o")
plt.grid()
plt.show()
```

```python
# 문제와 정답 데이터를 데이터프레임화

X_train = pd.DataFrame(X)
y_train = pd.DataFrame(y)
```

```python
# 모델 객체 생성

license_model = LinearRegression()
```

```python
# 모델 학습

license_model.fit(X_train, y_train)
```

```python
# 모델 예측

license_model.predict([[3]]) # 예상되는 예측값은 6

# 주의점 : 선형회귀 모델은 데이터를 2차원으로 넣어줘야 함,
#          1차원으로 데이터를 넣어줬을 시 에러 발생 [] -> [[]] 사용
```

```python
# 가중치와 편향값 확인
# 가중치 W값 확인 : coef_

print(license_model.coef_)
```

```python
# 편향(절편) B값 확인 : intercept_

print(license_model.intercept_)
```

##### 평균제곱오차 구하기

- 오차는 예측값에서 정답값을 뺀 값
- 각 오차의 제곱 구하기
